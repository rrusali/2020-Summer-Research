```python
from PIL import Image
```

```python
#####

# README: This code finds the average color of an image ignoring all pure black pixels

#####

im = Image.open('15213_roofs.png')
pix = im.load()
print(im.size)
```

```python
x, y = im.size
```

```python
color_pixels = []

# This creates an array with the RGB values of all non-black pixels

for i in range(x):
    
    for c in range(y):
        
        if pix[i, c] != (0, 0, 0, 255):
            
            value = [pix[i, c][0], pix[i, c][1], pix[i, c][2]]
            
            color_pixels.append(value)
            
color_pixels
```

```python
print(len(color_pixels))
```

```python
# Method for average color taken from: 
# https://sighack.com/post/averaging-rgb-colors-the-right-way#:~:text=The%20typical%20approach%20to%20averaging,color%2C%20sum%20their%20squares%20instead.

average_color = [0, 0, 0]

for i in range(3):
    
    square_sum = 0
    
    for c in range(len(color_pixels)):
        
        square_sum += (color_pixels[c][i])**2
        
    treated_color = (square_sum/len(color_pixels))**(1/2)
    
    average_color[i] = treated_color
    
print('The RGB code for the average color of you file is ' + str(average_color))
```
